 **Bivariate Functions**

#**3Dimensional Plot: P1 Maximum of F1**

```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
  return (-x ** 2 - y ** 2 + 5)

x = np.linspace(-20, 20, 20)
y = np.linspace(-20, 20, 20)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the figure
fig = plt.figure()
ax = plt.axes(projection='3d')

# Show graph
ax.view_init(46, 60)   #el,az
ax.contour3D(X, Y, Z, 70, cmap='Greys')
ax.plot([0.], [0.], [5.], markerfacecolor='red', markeredgecolor='k', marker='o', markersize=10, alpha=0.9, zorder=100)

# Label axis 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

# **Countor Plot: P1 Maximum of F1**

```
x = np.linspace(-10, 10, 500)
y = np.linspace(-10, 10, 500)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contour(X, Y, Z, colors='black');
```

# **Density Plot: P1 Maximum of F1**

```
plt.contourf(X, Y, Z, 20, cmap='RdGy');
plt.colorbar();
```
